*Utilizando a baseProcessada.csv obtida na atividade Formativa, você deve listar todas as variáveis que possuem comportamento comum, por exemplo: total day minutes, total eve minutes, total night minutes e total intl minutes. Estas variáveis representam o total de minutos falados em cada momento do dia e podem ser agrupadas em uma nova variável chamada TOTAL MINUTES. Para realizar este tipo de operação utilizando o pandas, os comandos são simples, como o exemplo a seguir:*

<pre>
df[‘TOTAL MINUTES’] = df[‘total day minutes’] + df[‘total eve minutes’] + df[‘total night minutes’] + df[‘total intl minutes’]
</pre>

Não se esqueça de listar TODAS as variáveis que possuem comportamentos comuns, e não apenas as referentes aos minutos.

```python
import pandas as pd
import numpy as np
import pandas_profiling
```

```python
churn = pd.read_csv('baseProcessada.csv')
#.transpose para facilitar a visualização
churn.head().T
```

```python
#renomeando nome das colunas não gosto de trabalhar com colunas com espaço
churn.columns = churn.columns.\
    str.replace('\s','_').\
    str.lower()
```

```python
churn.set_index('unnamed:_0', inplace=True)
churn.columns
```

## Transformando as colunas

```python
total_minutes = churn.filter(regex="minutes")
total_charge = churn.filter(regex='charge')
#usando re porque tem a coluna customer_service_calls
total_calls = churn.filter(regex='total_\w+_calls')
```

```python
churn['total_minutes'] = total_minutes.sum(axis=1)
churn['total_charge'] = total_charge.sum(axis=1)
churn['total_calls'] = total_calls.sum(axis=1)
```

## Removendo as colunas e mostrando como ficou o final

```python
columns_to_remove = []
columns_to_remove = total_minutes.columns.to_list() +  total_calls.columns.to_list() + total_charge.columns.to_list()
print(columns_to_remove)
churn_final = churn.drop(labels=columns_to_remove, axis=1)
churn_final.to_csv('baseProcessadaFinal.csv')
churn_final.head().T
```
